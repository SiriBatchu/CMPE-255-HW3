```python
import pycaret
pycaret.__version__
```

```python
### loading sample dataset from pycaret dataset module
from pycaret.datasets import get_data
data=get_data('uschange')
```

```python
# plot the dataset
data.plot()
```

```python
# import pycaret time series and init setup
from pycaret.time_series import *
s = setup(data,target='Consumption', session_id = 123)
```

```python
# import TSForecastingExperiment and init the class
from pycaret.time_series import TSForecastingExperiment
exp = TSForecastingExperiment()
```

```python
# check the type of exp
type(exp)
```

```python
# init setup on exp
exp.setup(data, target='Consumption', session_id = 123)
```

```python
# check statistical tests on original data
check_stats()
```

```python
# compare baseline models
best = compare_models()
```

```python
# compare models using OOP
exp.compare_models()
```

```python
# plot forecast
plot_model(best, plot = 'forecast')
```

```python
# residuals plot
plot_model(best, plot = 'residuals')
```

```python
# check docstring to see available plots
# help(plot_model)
```

```python
# predict on test set
holdout_pred = predict_model(best)
```
